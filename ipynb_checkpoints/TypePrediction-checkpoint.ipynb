```python
from sklearn.model_selection import train_test_split

X = dataset.drop(columns='Diabetic', axis=1)  # All the data except the Diabetic Column
Y = dataset['Diabetic']
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import r2_score

di = {}
di2 = {}
Knn = KNeighborsClassifier(n_neighbors=30)
Knn.fit(Xtrain,Ytrain)
Ypred = Knn.predict(Xtest)
err = metrics.mean_absolute_error(Ytest,Ypred)
di2.update({str(Knn):err})
print(Knn,":",r2_score(Ypred,Ytest)*100)
di.update({str(Knn):Knn.score(Xtest,Ytest)*100})
test_scores = []
train_scores = []
for i in range(5,30):
  neigh = KNeighborsClassifier(n_neighbors=i)
  neigh.fit(Xtrain,Ytrain)
  train_scores.append(neigh.score(Xtrain,Ytrain))
  test_scores.append(neigh.score(Xtest,Ytest))
print("Max train score is :",str(max(train_scores)*100), 'for k =', str(train_scores.index(max(train_scores))+5))
print("Max test score is :",str(max(test_scores)*100), 'for k =', str(test_scores.index(max(test_scores))+5))
```

```python
di = {}
di2 = {}
Knn = KNeighborsClassifier(n_neighbors=30)
Knn.fit(Xtrain,Ytrain)
Ypred = Knn.predict(Xtest)
err = metrics.mean_absolute_error(Ytest,Ypred)
di2.update({str(Knn):err})
print(Knn,":",r2_score(Ypred,Ytest)*100)
di.update({str(Knn):Knn.score(Xtest,Ytest)*100})
test_scores = []
train_scores = []
for i in range(5,30):
  neigh = KNeighborsClassifier(n_neighbors=i)
  neigh.fit(Xtrain,Ytrain)
  train_scores.append(neigh.score(Xtrain,Ytrain))
  test_scores.append(neigh.score(Xtest,Ytest))
print("Max train score is :",str(max(train_scores)*100), 'for k =', str(train_scores.index(max(train_scores))+5))
print("Max test score is :",str(max(test_scores)*100), 'for k =', str(test_scores.index(max(test_scores))+5))
```
